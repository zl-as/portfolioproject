```python
import pandas as pd
```

```python
dataframe2= pd.read_excel('Corona.xlsx')
print(dataframe2)
```

```python
dataframe1=dataframe2.dropna()
dataframe1
```

```python
import numpy as np

import matplotlib.pyplot as plt
dataframe1.plot(kind = 'bar')
```

```python
#Simple Feature scaling
# copy the data
df_scaled = dataframe1.copy()
  
# apply normalization techniques
for column in df_scaled.columns:
    if column == 'cases' or column == 'active cases'or column =='deaths':
          df_scaled[column] = df_scaled[column]  / df_scaled[column].max()
      
# view normalized data
display(df_scaled)
```

```python
#Min-max feature scaling

dataframe2 = dataframe1.copy()
  
# apply normalization techniques
for column in dataframe2.columns:
    if column == 'cases' or column == 'active cases'or column =='deaths':
        dataframe2[column] = (dataframe2[column] - dataframe2[column].min()) / (dataframe2[column].max() - dataframe2[column].min())    
  
# view normalized data
print(dataframe2)
```

```python
#z-score

# copy the data
df_z_scaled = dataframe1.copy()
  
# apply normalization techniques
for column in df_z_scaled.columns:
    if column == 'cases' or column == 'active cases'or column =='deaths':
        df_z_scaled[column] = (df_z_scaled[column] -
                           df_z_scaled[column].mean()) / df_z_scaled[column].std()    
  
#view normalized data   
display(df_z_scaled)
```
